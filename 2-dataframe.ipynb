```python
dados_notas = {
    'nome':['alice','bob','charlie','david','eva'],
    'matematica':[58,90,78,88,92],
    'fisica':[80,75,88,85,95],
    'quimica':[78,82,85,80,88]
}
```

```python
df_notas = pd.DataFrame(dados_notas)
df_notas
```

```python
#calcular a méddia das notas por aluno 
df_notas['Média'] = df_notas[["matematica","fisica","fisica"]].mean(axis=1)
df_notas
```

```python
#identificar alunos aprovados com nota excelente ( acuma de 83)
alunos_aprovados = df_notas[df_notas["Média"] >= 83]
alunos_aprovados
```

```python
#Exibir a média e os alunos aprovados 
df_notas[["nome",'Média']]
```
