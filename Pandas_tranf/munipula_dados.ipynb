```python
import pandas as pd
```

```python
df = pd.read_json("./files/dados_hospedagem.json")
df.head()
```

```python
df = pd.json_normalize(df["info_moveis"])
df.head()
```

```python
columns = list(df.columns)
columns
```

```python
df = df.explode(columns[3:])
df
```

```python
df.reset_index(inplace=True,drop=True)
df.head()
```

```python
df.dtypes
```

Conversão dados numéricos 

```python
import numpy as np
```

```python
df["max_hospedes"] = df["max_hospedes"].astype(np.int64)
```

```python
df.dtypes
```

```python
columns_numericas = ["quantidade_banheiros","quantidade_quartos","quantidade_camas"]
```

```python
df[columns_numericas] = df[columns_numericas].astype(np.int64)
df.dtypes
```

```python
df["avaliacao_geral"]
```

```python
df["avaliacao_geral"] = df["avaliacao_geral"].astype(np.float64)
df.dtypes
```
