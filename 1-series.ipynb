```python
#vendas mensais de uma loja 
vendas_mensais = pd.Series(
    [200,2500,3000,2800,3500,4000,4500,5000,4800,5200,6000,7000],
    index=["jan",'fev','mar','abr','mai','jun','jul','ago','set','out','nov','dez']
)
```

```python
vendas_mensais
```

```python
#calcular e exibir a média de vendas 
media_vendas = vendas_mensais.mean()
media_vendas
```

```python
#identificando mes com maor venda 
mes_maior_venda = vendas_mensais.idxmax()
mes_maior_venda
```

```python
#exibir vendas acima da média 
vendas_acima_media = vendas_mensais[vendas_mensais > media_vendas]
vendas_acima_media
```
