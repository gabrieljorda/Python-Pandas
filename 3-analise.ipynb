```python
import pandas as pd
```

```python
dados_vendas = {
    'produto':["Camisa",'Calça','Saia',"Camisa",'Calça',"Saia",'Camiseta','Calça','Saia'],
    'preco':[50,100,80,55,110,85,45,95,90],
    "quantidade":[10,15,20,5,10,15,30,20,10],
    "mes":["jan","jan",'jan','fev','fev','fev','mar','mar','mar']
}
```

```python
df_vendas = pd.DataFrame(dados_vendas)
df_vendas
```

```python
#Calculando a receira total
df_vendas["receita" ] = df_vendas["preco"] * df_vendas["quantidade"]
df_vendas
```

```python
#sumarizando a receita total por produto
resumo_vendas =  df_vendas.groupby("produto")["receita"].sum().reset_index()
resumo_vendas
```

```python
#Filtrar produtos com receita acima de 1000R$
produtos_altas_vendas = resumo_vendas[resumo_vendas["receita"]>1500]
produtos_altas_vendas
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,5))
plt.bar(resumo_vendas["produto"],resumo_vendas["receita"], color='skyblue')
plt.title('Receita Total por Produto')
plt.xlabel("produto")
plt.ylabel("Receita R$:")
plt.xticks(rotation=45)
```
