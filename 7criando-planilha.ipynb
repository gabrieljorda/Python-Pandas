```python
import pandas as pd
```

```python
dados = {
    'Nome': ['flamengo', 'santos', 'palmeiras','sao paulo', 'corinthians'],
    'partidas_jogadas': [10,10,10,10,10],
    'vitorias': [8,6,7,4,3] ,
    'empates': [1,2,2,3,4],
    'derrotas': [1,2,1,3,3],
    'gols_pro': [25,20,22,15,10],
    'gols_contra': [10,15,12,18,5]
}
```

```python
df = pd.DataFrame(dados)
df
```

```python
df['saldo_gols'] = df['gols_pro'] - df['gols_contra']
df
```

```python
df['pontos'] = df['vitorias'] * 3 + df['empates'] * 1
df
```

```python
df = df.sort_values(by='pontos', ascending=False)
df
```

```python
df.to_excel('dados/tabela_classificacao.xlsx', index=False)

```
